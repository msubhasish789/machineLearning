```python
import pandas as pd
```

```python
file_path = "Translation_Long.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
if "reference" not in df.columns or "translation" not in df.columns:
    raise ValueError("CSV file must contain 'reference' and 'translation' columns.")
```

```python
# Convert the 'translation' and 'reference' columns to strings before calculating the score
df["translation"] = df["translation"].astype(str)
df["reference"] = df["reference"].astype(str)

# Now calculate the score
P, R, F1 = score(df["translation"].tolist(), df["reference"].tolist(), lang="en", model_type="bert-base-uncased")
```

```python
F1
```

```python
df["BERTScore_F1"] = F1.tolist()
```

```python
output_file = "bert_score_long_results.csv"
df.to_csv(output_file, index=False)
```

```python
from bart_score import BARTScorer
```

```python
import pandas as pd
```

```python
import torch
```

```python
# Load pre-trained BART model
device = "cuda" if torch.cuda.is_available() else "cpu"
bart_scorer = BARTScorer(device=device, checkpoint="facebook/bart-large-cnn")
```

```python
file_path = "Translation_Long.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
if "reference" not in df.columns or "translation" not in df.columns:
    raise ValueError("CSV must contain 'reference' and 'translation' columns.")
```

```python
# Convert the 'translation' and 'reference' columns to strings before calculating the score
df["translation"] = df["translation"].astype(str)
df["reference"] = df["reference"].astype(str)
```

```python
df["BARTScore"] = bart_scorer.score(df["translation"].tolist(), df["reference"].tolist(), batch_size=4)
```

```python
df["BARTScore"]
```

```python
# Normalize BARTScore to a 0-100 range
df["BARTScore_Scaled"] = (df["BARTScore"] - df["BARTScore"].min()) / (df["BARTScore"].max() - df["BARTScore"].min()) * 100
```

```python
output_file = "bart_score_long_results.csv"
df.to_csv(output_file, index=False)
```

```python
print(f"Results saved to {output_file}")
```
